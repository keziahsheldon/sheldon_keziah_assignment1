```python
#Keziah Sheldon
#Problem 5
import numpy as np
import matplotlib.pyplot as plt
#import seaborn as sns

#Setting constants here
x=1
delta=1e-3
value=[]
increment=[]

#Calculates the f_prime at the different del values	 	
while delta > 1e-15:
    
    f_prime=( (x+delta)*((x + delta)-1.) - x*(x-1.) )/delta
    delta=delta*.25 #just getting smaller here
    
    value.append(f_prime)
    increment.append(delta)
    
plt.ylabel('Numerical output')
plt.xlabel('Value of del')
plt.title('Plot of the numerical accuracy at x=1')
plt.semilogx(increment,value,linestyle=":",marker="o")
plt.show()
#Oddly enough, the plot shows that at a certain point (around 1e-13) it starts to get wonky. 
#Is it because it is a float and is rounding off or has a certain limit to the way the information is stored?
```
